### Prediciton Model IO Linear Regresion 

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

### 50 Najlepszych dru≈ºyn

```python
best50 = pd.read_csv('best50.csv')
print(best50.shape)
best50.head()
```

```python
#przypis ID
best50['TeamId'] = [(x+1) for x in range(50)]

print(best50.shape)
best50.head()
```

```python
best50 = best50[['Team', 'TeamId']]
print(best50.shape)
best50.head()
```

### Wszystkie dane z Igrzysk Olimpijskich

```python
df = pd.read_csv('medalscountSUM.csv')
print(df.shape)
df.head()
```

```python
# Adding country_id column to summer df
df = pd.merge(df,best50, how = 'left', on = 'Team')
print(df.shape)
df.head()
```

```python
# Dropping the countries that are not in top 25
df = df.dropna()
print(df.shape)
df.head()
```

```python
df['Athletes - sport'] = round(df['Athletes']/df['Sports'], 2)
print(df.shape)
df.head()
```

```python
# Bring 'CountryId' column to the front
col = df.pop('TeamId')
df.insert(0,'TeamId', col)

print(df.shape)
df.head()
```

```python
df = df.sort_values(['Year', 'Medals', 'Team'], ascending = [True, False, True])
print(df.shape)
df.head()
```

```python
df = pd.get_dummies(df, columns = ['Team'])
print(df.shape)
df.head()
```

### Predykcja 2016

```python
predict_year = 2016
```

#Dane treningowe

```python
train_df = df[df['Year'] < predict_year]
print(train_df.shape)
train_df.head()
```

```python
X = train_df.drop(['Gold', 'Silver', 'Bronze', 'Medals'], axis=1)
```

```python
y1 = train_df['Gold'].values.reshape(-1, 1)
y2 = train_df['Silver'].values.reshape(-1, 1)
y3 = train_df['Bronze'].values.reshape(-1, 1)
y4 = train_df['Medals'].values.reshape(-1, 1)
```

```python
print(X.shape, y1.shape, y2.shape, y3.shape, y4.shape)
```

#Testowe dane 

```python
test_df = df[df['Year'] == predict_year].sort_values(['TeamId'])
print(test_df.shape)
test_df.head()
```

```python
test_data = test_df.drop(['Gold', 'Silver', 'Bronze', 'Medals'], axis = 1).reset_index(drop=True)
print(test_data.shape)
test_data.head()
```

```python
best50['Gold Actual'] = test_df['Gold'].reset_index(drop=True)
best50['Silver Actual'] = test_df['Silver'].reset_index(drop=True)
best50['Bronze Actual'] = test_df['Bronze'].reset_index(drop=True)
best50['All Medals Actual'] = test_df['Medals'].reset_index(drop=True)
```

```python
print(best50.shape)
best50.head()
```

#Test train split 

```python
from sklearn.model_selection import train_test_split
```

```python
#Gold
X1_train, X1_test, y1_train, y1_test = train_test_split(X, y1, random_state=2)
```

```python
#Silver
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y2, random_state=2)
```

```python
#Bronze
X3_train, X3_test, y3_train, y3_test = train_test_split(X, y3, random_state=2)
```

```python
#All Medals
X4_train, X4_test, y4_train, y4_test = train_test_split(X, y4, random_state=2)
```

```python
print(X1_train.shape, y1_train.shape, X1_test.shape,  y1_test.shape)
print(X2_train.shape, y2_train.shape, X2_test.shape,  y2_test.shape)
print(X3_train.shape, y3_train.shape, X3_test.shape,  y3_test.shape)
print(X4_train.shape, y4_train.shape, X4_test.shape,  y4_test.shape)
```

Regresja liniowa

```python
from sklearn.linear_model import LinearRegression
```

```python
training_score1 = model1.score(X1_train, y1_train)
testing_score1 = model1.score(X1_test, y1_test)
```

```python
print('Gold Medals:')
print(f"Gold Training Score: {training_score1}")
print(f"Gold Testing Score: {testing_score1}")
```
